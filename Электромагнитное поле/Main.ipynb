```python
import math
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import time

def PLOT2(X,Y,s):
    fig = plt.figure()
    plt.grid()
    plt.scatter(0,0, color="red", s=100)
    plt.xlabel(s[0])
    plt.ylabel(s[1])
    #plt.xlim(-2,2)
    #plt.ylim(-2,2)
    plt.plot(X,Y)

def PLOT(X,Y,Z):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    plt.scatter(0,0, color="red", s=100)
    plt.xlabel("x")
    plt.ylabel("y")
    ax.scatter(X,Y,Z)
    plt.show()
    PLOT2(X,Y,"xy")
    PLOT2(X,Z,"xz")
    PLOT2(Z,Y,"zy")
```

```python
#Электрическое поле
x0 = 0
y0 = 0
z0 = 0
V0x = 0.5
V0y = 0.4
V0z = 0.3
Ex = 2
Ey = 2
Ez = 2
m = 0.1
q = -0.1

n = 100
t0 = 0
tn = 1
h = (tn-t0)/n

X = [x0]
Y = [y0]
Z = [z0]
t = t0+h
for i in range(1,n):
    X.append(x0+V0x*t+(q*Ex*t**2)/(2*m))
    Y.append(y0+V0y*t+(q*Ey*t**2)/(2*m))
    Z.append(z0+V0z*t+(q*Ez*t**2)/(2*m))
    t+=h
    
PLOT(X,Y,Z)
```

```python
#Магнитное поле
B = 0.2
m = 0.1
q = 0.1
V = 1
Vz = 2
W = q*B/m
p = V/W
x0 = 0
y0 = 0
z0 = 0

n = 1000
t0 = 0
tn = 100
h = (tn-t0)/n

X = [x0]
Y = [y0]
Z = [z0]
t = t0+h
for i in range(1,n):
    X.append(x0-p*math.cos(W*t))
    Y.append(y0+p*math.sin(W*t))
    Z.append(z0+Vz*t)
    t+=h
    
PLOT(X,Y,Z)
```

```python
#Электрическое и магнитное поля
alpha = 0
E = 2
B = 0.2

x0 = 0
y0 = 0
z0 = 0
V0x = 0.1
V0y = 0.4
V0z = 0.3
Ez = E*math.cos(alpha*math.pi/180)
Ex = E*math.sin(alpha*math.pi/180)
m = 0.1
q = 0.1

W = q*B/m
p = V/W

n = 1000
t0 = 0
tn = 100
h = (tn-t0)/n

X = [x0]
Y = [y0]
Z = [z0]
t = t0+h
for i in range(1,n):
    X.append(x0+V0x*t-p*math.cos(W*t)+(q*Ex*t**2)/(2*m))
    Y.append(y0+V0y*t+p*math.sin(W*t))
    Z.append(z0+V0z*t+(q*Ez*t**2)/(2*m))
    t+=h
    
PLOT(X,Y,Z)
```
